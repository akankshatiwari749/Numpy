# Basic Image Manipulation using Numpy

The program should be menu-based with the following top-level features:
1. Load Image
2. Edit Image
3. Save Image
4. Quit


‘Edit Image’ should be able to do the following using numpy operations:
1. Horizontal Flip
2. Vertical Flip
3. Invert Colors
4. Convert to Grayscale

```python
print("Welcome to the mini-project")
t=1
while t:
    print("Enter your choices")
    print(" 1: for image loading \n 2: for editing \n 3: for saving\n 4: for exit")
    choice=int(input("Enter your choice:"))
    
    if choice == 1:
        np_image=io.imread('img1.jpg')
        plt.imshow(np_image)
        print("Image Loding is done\n")
        plt.show()
    elif choice == 2:
        print("\nHow you want to edit\n 1 for horizontal flip\n 2 for vertical flip \n 3 for invert colors\n 4 for grayscale\n ")
        fin=int(input("Enter your choice:"))
        if fin==1:
            func1()
            print("Image is horizontal fliped\n")
        elif fin==2:
            func2()
            print("Image is vertical fliped\n")
        elif fin==3:
            func3()
            print("Image color is inverted\n")
        elif fin==4:
            func4()
            print("Image color is conert to grayscale\n")
        else:
            print("Wrong choice")
            continue
    elif choice == 3:
        print("Image is saved\n")
    else:
        t=0
    
```
